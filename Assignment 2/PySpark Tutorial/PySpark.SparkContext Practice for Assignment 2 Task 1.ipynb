```python
import time

start = time.time()
distinct_mode_rdd = income_rdd.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b).sortByKey()
distinct = distinct_mode_rdd.count()
mode = distinct_mode_rdd.max(lambda x: x[1])[0]
med_loc = income_rdd.count()/2

median = 0
summ = 0
c = 0
lst = distinct_mode_rdd.collect()
while med_loc > summ:
    median = lst[c][0]
    summ += lst[c][1]
    c += 1

print("The Number of Distinct Values: {}".format(distinct))
print("Median: {}".format(median))
print("Mode: {}\n".format(mode))

count_10_rdd = distinct_mode_rdd.map(lambda x: (len(x[0])-1, x[1])).reduceByKey(lambda a, b: a+b)
count_10 = count_10_rdd.collect()
print("*****Count Per 10 Powers*****\n")
for (k, v) in count_10:
    print('10 ** {}: {}'.format(k, v))
print('\n',time.time() - start)
```

```python
test = sc.textFile('test_incomes.csv')
```

```python
import time

start = time.time()
distinct_mode_rdd = test.map(lambda x: (int(x), 1)).reduceByKey(lambda a, b: a + b).sortByKey()
distinct = distinct_mode_rdd.count()
mode = distinct_mode_rdd.max(lambda x: x[1])[0]
med_loc = test.count()/2

median = 0
summ = 0
c = 0
lst = distinct_mode_rdd.collect()
while med_loc > summ:
    median = lst[c][0]
    summ += lst[c][1]
    c += 1

print("The Number of Distinct Values: {}".format(distinct))
print("Median: {}".format(median))
print("Mode: {}\n".format(mode))

count_10_rdd = distinct_mode_rdd.map(lambda x: (len(str(x[0]))-1, x[1])).reduceByKey(lambda a, b: a+b)
count_10 = count_10_rdd.collect()
print("*****Count Per 10 Powers*****\n")
for (k, v) in count_10:
    print('10 ** {}: {}'.format(k, v))
print('\n',time.time() - start)
```
